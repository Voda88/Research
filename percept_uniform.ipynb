# 2

We can prove this using the basic theorems of probability

\begin{align}
P(A+B\!\mid\!I)&=\frac{P((A\text{ or }B)\text{ and }I)}{P(I)}\\
&=\frac{P(A\text{ and }I\text{ or }B\text{ and }I)}{P(I)}\\
&=\frac{P(A\text{ and }I)+P(B\text{ and }I)-P((A\text{ and }I)\text{ and }P(B\text{ and }I))}{P(I)}\\
&=\frac{P(A\text{ and }I)}{P(I)}+\frac{P(B\text{ and }I)}{P(I)}-\frac{P((A\text{ and }B)\text{ and }I)}{P(I)}\\
&=P(A\!\mid\!I)+P(B\!\mid\!I)-P(A,B\!\mid\!I)
\end{align}

The cell below shows the iterative formula. However, this is inefficient, so to calculate $C_{113,59}$, we employ the &nbsp;&nbsp;```scipy.misc.comb```&nbsp;&nbsp; function; the answer is about $69.69\times10^{31}$

````python
def choose(n,k):
    if k>n:
        return 0
    if k==n or k==0:
        return 1
    else:
        return choose(n-1,k)+choose(n-1,k-1)
````

The priors on $x$ and $y$ mean that they're normalized over the unit square. To get the marginal distributions, we integrate away the other variable over the range $[0,1]$.

\begin{align}
p(x\,|\,I)&=\int\limits_{y\in[0,1]}{p(x,y\,|\,I)}\,dy\\
&=\int_{0}^{1}\left[\,A\exp\left(-\frac{(x-x_1)^2+(y-y_1)^2}{2{\sigma_1}^2}\right)+B\exp\left(-\frac{(x-x_2)^2+(y-y_2)^2}{2{\sigma_2}^2}\right)\right]\,dy\\
&=A\exp\left(-\frac{(x-x_1)^2}{2{\sigma_1}^2}\right)\int_{0}^{1}\exp\left(-\frac{(y-y_1)^2}{2{\sigma_1}^2}\right)\,dy+B\exp\left(-\frac{(x-x_2)^2}{2{\sigma_2}^2}\right)\int_{0}^{1}\exp\left(-\frac{(y-y_2)^2}{2{\sigma_2}^2}\right)\,dy\\
&={A\exp\left(-\frac{(x-x_1)^2}{2{\sigma_1}^2}\right)}{\left[\sigma_1\sqrt\frac{\pi}{2}\left(\operatorname{erf}\left(\frac{y_1}{\sigma_1\sqrt2}\right)-\operatorname{erf}\left(\frac{y_1-1}{\sigma_1\sqrt2}\right)\right)\right]}+{B\exp\left(-\frac{(x-x_2)^2}{2{\sigma_2}^2}\right)}{\left[\sigma_2\sqrt\frac{\pi}{2}\left(\operatorname{erf}\left(\frac{y_2}{\sigma_2\sqrt2}\right)-\operatorname{erf}\left(\frac{y_2-1}{\sigma_2\sqrt2}\right)\right)\right]}\\
p(y\,|\,I)&=\int\limits_{x\in[0,1]}{p(x,y\,|\,I)}\,dx\\
&=\int_{0}^{1}\left[\,A\exp\left(-\frac{(y-y_1)^2+(x-x_1)^2}{2{\sigma_1}^2}\right)+B\exp\left(-\frac{(y-y_2)^2+(x-x_2)^2}{2{\sigma_2}^2}\right)\right]\,dx\\
&=A\exp\left(-\frac{(y-y_1)^2}{2{\sigma_1}^2}\right)\int_{0}^{1}\exp\left(-\frac{(x-x_1)^2}{2{\sigma_1}^2}\right)\,dx+B\exp\left(-\frac{(y-y_2)^2}{2{\sigma_2}^2}\right)\int_{0}^{1}\exp\left(-\frac{(x-x_2)^2}{2{\sigma_2}^2}\right)\,dx\\
&={A\exp\left(-\frac{(y-y_1)^2}{2{\sigma_1}^2}\right)}{\left[\sigma_1\sqrt\frac{\pi}{2}\left(\operatorname{erf}\left(\frac{x_1}{\sigma_1\sqrt2}\right)-\operatorname{erf}\left(\frac{x_1-1}{\sigma_1\sqrt2}\right)\right)\right]}+{B\exp\left(-\frac{(y-y_2)^2}{2{\sigma_2}^2}\right)}{\left[\sigma_2\sqrt\frac{\pi}{2}\left(\operatorname{erf}\left(\frac{x_2}{\sigma_2\sqrt2}\right)-\operatorname{erf}\left(\frac{x_2-1}{\sigma_2\sqrt2}\right)\right)\right]}\\
\end{align}

````python
def p(x,y,A,B,x1,y1,x2,y2,s1,s2):
    return A*np.exp(-((x-x1)**2+(y-y1)**2)/(2*s1**2))+B*np.exp(-((x-x2)**2+(y-y2)**2)/(2*s2**2))
````

````python
import matplotlib.cm as cm
````

````python
x,y = np.meshgrid(np.linspace(0,1,501),np.linspace(0,1,501))
````

````python
z=p(x,y,3.93134,3.61447,0.5,0.5,0.65,0.75,0.2,0.04)
````

````python
plt.figure(figsize=(10,10))
im = plt.imshow(z[::-1],cmap=cm.magma,interpolation="bilinear",extent=[0,1,0,1],vmin=0,vmax=5)
cbar = plt.colorbar(im,fraction=0.046,pad=0.04,label='Magnitude')
cset = plt.contour(z,np.arange(0,6,0.5),linewidths=.8,cmap=cm.viridis_r,extent=[0,1,0,1])
plt.clabel(cset,inline=True,fmt="%1.1f",fontsize=8)
plt.xlabel('x position')
plt.ylabel('y position')
plt.title('Color plot and contour plot of two Gaussian sources')
plt.show()
````

````python
from scipy.special import erf
````

````python
def px(x,A,B,x1,y1,x2,y2,s1,s2):
    return A*np.exp(-(x-x1)**2/(2*s1**2))*s1*(np.pi/2)**.5*(erf(y1/(s1*2**.5))-erf((y1-1)/(s1*2**.5)))+\
           B*np.exp(-(x-x2)**2/(2*s2**2))*s2*(np.pi/2)**.5*(erf(y2/(s2*2**.5))-erf((y2-1)/(s2*2**.5)))

def py(y,A,B,x1,y1,x2,y2,s1,s2):
    return A*np.exp(-(y-y1)**2/(2*s1**2))*s1*(np.pi/2)**.5*(erf(x1/(s1*2**.5))-erf((x1-1)/(s1*2**.5)))+\
           B*np.exp(-(y-y2)**2/(2*s2**2))*s2*(np.pi/2)**.5*(erf(x2/(s2*2**.5))-erf((x2-1)/(s2*2**.5)))
````

````python
left, width = 0.1,0.65
bottom, height = 0.1,0.65
bottom_h = left_h = left + width + 0.07
cbarHeight = 0.03

rect_color = [left, bottom, width, height]
rect_histx = [left, bottom_h + 0.02, width, 0.15]
rect_histy = [left_h, bottom, 0.15, height]

fig = plt.figure(figsize=(9,9))

axColor = fig.add_axes(rect_color)
im = plt.imshow(z[::-1],cmap=cm.magma,interpolation="bicubic",extent=[0,1,0,1],vmin=0,vmax=5)
cset = plt.contour(z,np.arange(0,6,0.5),linewidths=.8,cmap=cm.viridis_r,extent=[0,1,0,1])
axColor.clabel(cset,inline=True,fmt="%1.1f",fontsize=8)
axColor.set(xlabel='x position',ylabel='y position',title='Color plot and contour plot of two Gaussian sources')
axColor.title.set_position([0.5,1.01])

histBin = np.linspace(0,1,10001)
axHistx = fig.add_axes(rect_histx)
axHistx.plot(histBin,px(histBin,3.93134,3.61447,0.5,0.5,0.65,0.75,0.2,0.04),'b-',lw=1)
axHistx.set(xlim=[0,1],ylabel='$p(x\,|\,I)$',title='Marginal x distribution')
axHistx.title.set_position([0.5,1.05])

axHisty = fig.add_axes(rect_histy)
axHisty.plot(py(histBin,3.93134,3.61447,0.5,0.5,0.65,0.75,0.2,0.04),histBin,'b-',lw=1)
axHisty.set(ylim=[0,1],xlabel='$p(y\,|\,I)$',title='Marginal y\ndistribution')
axHisty.title.set_position([0.5,1.015])

axCbar = fig.add_axes([left,0, width, cbarHeight])
cbar = plt.colorbar(im,cax=axCbar,label='Magnitude',orientation='horizontal')
cbar.set_ticks(np.arange(0,5.1,0.5))

plt.savefig('gaussians.eps',bbox_inches='tight')
plt.savefig('gaussians.png',bbox_inches='tight',dpi=300)
````
